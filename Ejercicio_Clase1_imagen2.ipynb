```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
W=200
H=200
```

```python
# Crear una imagen negra
im = np.zeros((W, H, 3), np.uint8)
```

```python
# Definir las coordenadas del cuadro de fondo negro
x1, y1, x2, y2 = 50, 50, 150, 150
```

```python
# Lista de colores y tama√±os de los cuadrados
cuadrados = [((0, 0, 255), 5), ((0, 255, 0), 15), ((255, 0, 0), 25), ((0, 0, 0), 35)]
```

```python
# Dibujar cuadrados con un bucle
for color, size in cuadrados:
    im[y1 + size:y2 - size, x1 + size:x2 - size] = color

# Mostrar la imagen resultante
cv2_imshow(im)
```
