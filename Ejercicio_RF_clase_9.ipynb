```python
file_path = '/content/drive/MyDrive/Visión por computadora/Tareas/Actividad_Clase_9/mushrooms.csv'
df = pd.read_csv(file_path)
```

```python
#Preprocesar los datos
from sklearn.preprocessing import LabelEncoder
```

```python
#Convertir las características categóricas a numéricas usando LabelEncoder
label_encoder = LabelEncoder()
for column in df.columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
#Definir las características (X) y la etiqueta (y)
X = df.drop('class', axis=1)
y = df['class']
```

```python
#Dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Entrenar el modelo de árbol de decisión
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
#Evaluar el modelo
from sklearn.metrics import classification_report, accuracy_score

y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
